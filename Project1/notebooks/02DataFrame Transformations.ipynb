```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

spark = SparkSession.builder.getOrCreate()

data = [
    (1, "Alice", 25, "HR", 50000),
    (2, "Bob", 30, "IT", 70000),
    (3, "Charlie", 35, "IT", 90000),
    (4, "David", 28, "Finance", 65000),
    (5, "Eva", 22, "HR", 45000),
    (6, "Frank", 30, "IT", 70000)
]

columns = ["id", "name", "age", "department", "salary"]

df = spark.createDataFrame(data, columns)

df.show()
```

```python
df.select("name", "salary").show()
```

```python
df.select(col("name"), col("salary")).show()
```

```python
df.select(col("name").alias("Employee Name"), col("salary").alias("Inhand salary")).show()
```

```python
df_new = df.withColumn(
    "new_salary",
    round(col("salary") * 1.10, 2)
)

df_new.show()
```

```python
df_new = (
    df
    .withColumn("bonus", col("salary") * 0.15)
    .withColumn("total_salary", col("salary") + col("bonus"))
)

df_new.show()
```

```python
df.withColumnRenamed("department", "dept").show()
```

```python
df = (
    df
    .withColumnRenamed("department", "dept")
    .withColumnRenamed("salary", "income")
)
```

```python
df.show()
```

```python
df.drop("age").show()
```

```python
df.where(col("salary") > 60000).show()
```

```python
df.where((col('dept') == "IT") &
         (col('age') > 30) &
         (col('income') > 50000)).show()
```

```python
df.where(col("salary") > 60000).orderBy(desc("income")).show()
```

```python
df.distinct().show()
```

```python
df.dropDuplicates(["dept", "income"]).show()
```

```python
result = (
    df
    .filter(col("income") > 50000)
    .select("name", "dept", "income")
    .orderBy(desc("income"))
)

result.show()
```
